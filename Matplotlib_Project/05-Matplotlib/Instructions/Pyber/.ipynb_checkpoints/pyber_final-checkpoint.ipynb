```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_df = pd.merge(ride_data, city_data, how="left", on="city")

# Display the data table for preview
city_ride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types urban,rural, suburban

urban_type = city_ride_df[city_ride_df["type"] == "Urban"].groupby([city_ride_df["city"]])
rural_type = city_ride_df[city_ride_df["type"] == "Rural"].groupby([city_ride_df["city"]])
suburban_type = city_ride_df[city_ride_df["type"] == "Suburban"].groupby([city_ride_df["city"]])

# URBAN CALCULATIONS
# ride count (x)
x_urban = urban_type["ride_id"].count()      
 # fare average(y)
y_urban = urban_type["fare"].mean()         
# driver count average size(s)
s_urban = urban_type["driver_count"].mean() 

# RURAL CALCULATIONS 
# ride count (x)
x_rural = rural_type["ride_id"].count()       
# fare average(y)
y_rural = rural_type["fare"].mean()  
# driver count average size(s)
s_rural = rural_type["driver_count"].mean()   

# SUBURBAN CALCULATIONS
# ride count (x)
x_suburban = suburban_type["ride_id"].count()      
 # fare average(y)
y_suburban = suburban_type["fare"].mean()         
# driver count average size(s)
s_suburban = suburban_type["driver_count"].mean()


# Build the scatter plots for each city types
# URBAN Scatter Plot
plt.scatter(x_urban, 
            y_urban, 
            s = s_urban * 10, 
            marker="o", 
            facecolor = "lightcoral", 
            edgecolor = "black", 
            label = "Urban", 
            alpha = 0.75,
            linewidth = 0.75)
# RURAL Scatter Plot
plt.scatter(x_rural, 
            y_rural, 
            s = s_rural * 10, 
            marker ="o", 
            facecolor = "gold", 
            edgecolor = "black", 
            label = "Rural", 
            alpha = 0.75,
            linewidth = 0.75)
# SUBURBAN Scatter Plot
plt.scatter(x_suburban, 
            y_suburban, 
            s = s_suburban * 10, 
            marker ="o", 
            facecolor = "lightskyblue", 
            edgecolor = "black", 
            label = "Suburban", 
            alpha = 0.75,
            linewidth = 0.75)



# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
plt.axis("tight")


# Create a legend
legend = plt.legend(fontsize = 10, title = "City Types", edgecolor = "black", loc = "best")
legend.legendHandles[0]._sizes = [80]
legend.legendHandles[1]._sizes = [80]
legend.legendHandles[2]._sizes = [80]

# Incorporate a text label regarding circle size
plt.text(45,34,"Note:")
plt.text(45,32,"Circle size correlates with driver count per city.")

# Save Figure
plt.savefig("../Images/bubble_plot.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_df = city_ride_df.groupby(["type"])
type_percents = type_df["fare"].sum()

# Build Pie Chart
plt.pie(type_percents, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], 
        explode=[0, 0, 0.1], 
        autopct='%1.1f%%', 
        shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Images/total_fares.png")

# Show plot
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_percents = 100 * city_ride_df.groupby(["type"]).count()["ride_id"] / city_ride_df["ride_id"].count()

# Build Pie Chart
plt.pie(ride_percents, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], 
        explode=[0, 0, 0.1], 
        autopct='%1.1f%%', 
        shadow= True, startangle=100)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("../Images/total_rides.png")

# Show plot
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_percents = 100 * city_data.groupby(["type"]).sum()["driver_count"] / city_data["driver_count"].sum()

# Build Pie Charts
plt.pie(driver_percents, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], 
        explode=[0, 0, 0.1], 
        autopct='%1.1f%%', pctdistance=.5,
        shadow=True, startangle=120)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("../Images/total_drivers.png")

# Show plot
plt.show()
```
